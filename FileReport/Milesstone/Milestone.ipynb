## 5.2. Thông tin về project

***Bài toán nhóm muốn giải quyết***

Trong chủ đề này, nhóm sẽ giải quyết bài toán sử dụng **Twitter API** để truy cập và tương tác với dữ liệu từ Twitter, cụ thể là thu thập các tweet, thông tin người dùng, xu hướng (trends) và các dữ liệu khác liên quan đến hoạt động trên Twitter. Mục tiêu cuối cùng là thu thập và xử lý dữ liệu liên quan đến chủ đề Black Friday, thường là các bài quảng cáo, để xây dựng một mô hình học sâu (Deep Learning) với mạng nơ-ron hồi tiếp dài hạn (LSTM - Long Short-Term Memory).

***Tập dữ liệu*** 

**Dữ liệu thu nhập từ Twitter**

Dữ liệu nhận được từ Twitter là một JSON object chứa các thông tin chi tiết về một tweet bao gồm các siêu dữ liệu liên quan đến tweet, video và thông tin về người dùng.


***input của bài toán***

Cấu trúc dữ liệu bao gồm:

**Thông tin chung về Tweet:**

"tweet_id": ID duy nhất của tweet

"creation_date": Ngày và giờ tạo tweet

"text": Nội dung của tweet #sport

"language": Ngôn ngữ của tweet

"favorite_count": Số lượt thích tweet đã nhận 

"retweet_count": Số lượt retweet 

"reply_count": Số lượt trả lời 

"quote_count": Số lượt tweet đã được trích dẫn 

"views": Số lượt xem của tweet 

"expanded_url": URL đầy đủ của tweet 

"source": Nền tảng từ đó tweet được đăng tải

**Thông tin về Video và Media:**

"media_url": Danh sách URL của hình ảnh đại diện của video

"video_url": Danh sách các URL của các phiên bản video với các mức bitrate khác nhau:

URL m3u8 dành cho streaming

URL MP4 với các bitrate khác nhau như 632000, 950000, và 2176000.

"extended_entities": Thông tin chi tiết về media trong tweet:

"media": Danh sách các media đính kèm trong tweet


"type": Loại media (video).

"sizes": Các kích thước có sẵn của media (large, medium, small, thumb).

"video_info": Thông tin chi tiết về video như tỷ lệ (aspect_ratio), thời lượng (duration_millis), và các phiên bản video (variants) với bitrate khác nhau.

**Thông tin về người dùng:**

"user_id": ID của người dùng

"username": Tên đăng nhập của người dùng

"name": Tên hiển thị của người dùng

"follower_count": Số người theo dõi

"following_count": Số người dùng đang theo dõi 

"favourites_count": Số lượng tweet đã đánh dấu yêu thích

"is_private": Kiểm tra tài khoản có ở chế độ riêng tư (null có nghĩa là không có thông tin)

"is_verified": Người dùng có tick xanh xác thực không

"is_blue_verified": Người dùng có xác thực Blue Verified không 

"location": Vị trí của người dùng

"profile_pic_url": URL ảnh đại diện của người dùng

"profile_banner_url": URL banner của người dùng

"description": Phần mô tả của người dùng

"external_url": URL bên ngoài liên kết với hồ sơ của người dùng

"number_of_tweets": Số lượng tweet người dùng đã đăng

"bot": Xác định người dùng có phải bot không 

"timestamp": Thời điểm tạo tài khoản (dưới dạng UNIX timestamp)

**Các trường khác**

"retweet": Kiểm tra tweet có phải là retweet không

"in_reply_to_status_id": ID của tweet mà tweet này trả lời

"quoted_status_id": ID của tweet được trích dẫn

"binding_values": Các giá trị liên kết

"conversation_id": ID cuộc hội thoại chứa tweet này

"retweet_status": Trạng thái retweet

"quoted_status": Trạng thái trích dẫn

"bookmark_count": Số lượt đánh dấu của tweet 

"community_note": Ghi chú cộng đồn

**Dữ liệu để training model**

Sau khi thu thập được file JSON chứa dữ liệu Tweets, việc tiền xử lý là một bước quan trọng để chuẩn bị dữ liệu cho các bước tiếp theo. Trong trường hợp này, mục tiêu là **chỉ giữ lại nội dung của đoạn Tweets (thuộc tính “text”)** và chuyển đổi dữ liệu sang định dạng file .txt để dễ dàng sử dụng trong việc training mô hình.

***output của bài toán***
- Mục tiêu cuối cùng của bài toán này là xây dựng một mô hình LSTM (Long Short-Term Memory) có khả năng tự động sinh văn bản dựa trên các bài viết được thu thập từ Twitter liên quan đến Black Friday. Sau khi mô hình được huấn luyện với dữ liệu tweet, output của mô hình sẽ là khả năng tạo ra các bài viết (tweets) có nội dung tương tự, mang đậm tính chất của các chiến dịch quảng cáo hoặc khuyến mãi trong mùa Black Friday.

***Cách đánh giá kết quả***
- So sánh loss cuối cùng: Xem kết quả loss của epoch cuối cùng (epoch 63) và so với các epoch trước đó để đánh giá sự cải thiện. Nếu không có cải thiện đáng kể, có thể mô hình đã học đầy đủ các đặc trưng của dữ liệu.

## 5.3. Giải pháp

- Trong bài toán này, nhóm đã không sử dụng API chính thức của Twitter do chính sách thay đổi, yêu cầu phải trả thêm phí. Thay vào đó, nhóm đã lựa chọn **RapidAPI**, một nền tảng cung cấp API mạng xã hội cho người dùng, giúp truy cập vào dữ liệu từ Twitter mà không phải trả chi phí cao. Quá trình thu thập dữ liệu sẽ được tự động hóa thông qua **Apache NiFi**, một công cụ mạnh mẽ để xây dựng các workflow xử lý dữ liệu.

- Dữ liệu thu thập được sẽ được gửi tới **Apache Kafka** để thực hiện streaming, giúp duy trì sự liên kết giữa các hệ thống và đảm bảo tính sẵn sàng cao cho việc xử lý dữ liệu theo thời gian thực. Tại đây, dữ liệu sẽ được lưu trữ và truyền tải một cách linh hoạt, hỗ trợ việc truy xuất dễ dàng trong các giai đoạn sau của quy trình.

- Sau khi dữ liệu được thu thập, nó sẽ được xử lý và chuyển đổi định dạng từ JSON sang TXT để thuận tiện cho việc huấn luyện mô hình. Quá trình này sẽ được thực hiện bởi **Apache Spark**,các thuộc tính không cần thiết sẽ được loại bỏ chỉ giữ lại phần văn bản tweet (“text”), giúp dữ liệu sạch và dễ sử dụng cho các bước tiếp theo.

- Mô hình chính được nhóm sử dụng để xử lý dữ liệu này là LSTM (Long Short-Term Memory), một loại mạng nơ-ron hồi tiếp (RNN) đặc biệt hiệu quả trong các bài toán có tính chuỗi, như phân tích văn bản hay tạo ra văn bản tự động. Mô hình này sẽ giúp học được các mối quan hệ giữa các từ trong văn bản tweet và sử dụng kiến thức này để sinh ra các đoạn văn bản mới tương tự.

- Cấu trúc mô hình LSTM trong bài toán bao gồm hai lớp LSTM, mỗi lớp có 256 đơn vị, kết hợp với lớp Dropout để tránh overfitting và tăng khả năng tổng quát cho mô hình. Đầu ra của mô hình sẽ là một lớp **Dense** với hàm kích hoạt **softmax**, giúp dự đoán từ tiếp theo trong chuỗi văn bản, từ đó sinh ra các câu hoàn chỉnh.

- Quá trình huấn luyện mô hình sử dụng **categorical crossentropy** làm hàm mất mát, phù hợp với bài toán phân loại đa lớp. **Adam optimizer** sẽ được sử dụng để tối ưu hóa quá trình huấn luyện. Mô hình sẽ được lưu lại mỗi khi có sự cải thiện về loss, nhờ vào ModelCheckpoint.

Thuật toán này đặc biệt phù hợp trong bài toán **xử lý ngôn ngữ tự nhiên (NLP)**, nơi ngữ cảnh đóng vai trò quan trọng để hiểu và tạo ra chuỗi văn bản có nghĩa. Trong trường hợp này, LSTM không chỉ học cách dự đoán ký tự tiếp theo mà còn đảm bảo tính liên kết giữa các ký tự để sinh ra một văn bản hoàn chỉnh, tự nhiên.

## 5.4. Kết quả sơ bộ 

- Dự án này đã chứng minh tiềm năng mạnh mẽ của công nghệ Text Generation trong việc tự động hóa quy trình tạo và đăng nội dung quảng cáo trên mạng xã hội. Bằng cách tích hợp các công cụ tiên tiến như Apache NiFi, Apache Kafka, Apache Spark, và các mô hình học sâu như LSTM, hệ thống không chỉ giúp tiết kiệm thời gian mà còn tăng cường khả năng tiếp cận khách hàng một cách linh hoạt và hiệu quả. Dữ liệu thu thập từ Twitter thông qua API của RapidAPI được xử lý và chuyển đổi, sau đó được sử dụng để huấn luyện mô hình, giúp tạo ra các bài viết quảng cáo phù hợp, hỗ trợ tối ưu hóa chiến lược tiếp thị, tăng cường hiệu quả quảng cáo và tiết kiệm nguồn lực.
- Kết quả thực nghiệm cho thấy mô hình LSTM có khả năng tạo ra các bài đăng tự động với nội dung phù hợp và liên quan đến chủ đề Black Friday. Điều này không chỉ cải thiện quy trình sản xuất nội dung mà còn giúp giảm thiểu chi phí và thời gian cần thiết để tạo ra các bài viết quảng cáo.
- Tuy nhiên, dự án cũng đối mặt với một số thách thức quan trọng, bao gồm việc nâng cao chất lượng nội dung mà mô hình tạo ra. Mặc dù mô hình có thể tạo ra các bài viết tương đối mạch lạc, nhưng vẫn còn một số vấn đề cần cải thiện. Đặc biệt, việc đảm bảo tính chính xác, sáng tạo và tự nhiên trong các bài đăng vẫn là một thách thức. Mô hình đôi khi gặp khó khăn trong việc tạo ra các nội dung không chỉ hợp lý mà còn có tính ngữ nghĩa cao và dễ hiểu đối với người đọc. Cần có thêm các cải tiến trong việc tối ưu hóa mô hình, chẳng hạn như áp dụng các kỹ thuật điều chỉnh siêu tham số hoặc sử dụng các mô hình học sâu mạnh mẽ hơn như Transformer để nâng cao khả năng sinh ngữ nghĩa của văn bản.
